```python
# Load Spacy for Lemmatization
#Without Extra Features
import re
import pandas as pd
import numpy as np
import spacy
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTETomek
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from collections import Counter

# Load spaCy model (disable unnecessary components for speed)
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

# Optimized text cleaning function using spaCy
def clean_text_spacy(text):
    """ Use spaCy for fast lemmatization & text cleaning """
    text = re.sub(r"\W+", " ", text.lower())  # Remove special characters
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])  # Apply lemmatization

# Enable progress bars for better tracking
tqdm.pandas()

# Convert lists into DataFrames
df_train = pd.DataFrame(random_train_articles)
df_val = pd.DataFrame(random_val_articles)
df_test = pd.DataFrame(random_test_articles)

# Drop problematic sources
drop_list = {"Guest Writer", "Yellow Scene Magazine", "Aaron Carroll", "Ken Burns",
             "Carol Costello", "The Intelligencer", "Polish Times"}
df_train = df_train.loc[~df_train["source"].isin(drop_list)]
df_val = df_val.loc[~df_val["source"].isin(drop_list)]
df_test = df_test.loc[~df_test["source"].isin(drop_list)]

# Apply optimized text cleaning
df_train["content"] = df_train["content"].progress_apply(clean_text_spacy)
df_val["content"] = df_val["content"].progress_apply(clean_text_spacy)
df_test["content"] = df_test["content"].progress_apply(clean_text_spacy)

# Define text features & labels
X_train_text = df_train["content"]
X_val_text = df_val["content"]
X_test_text = df_test["content"]

y_train = df_train["bias"]
y_val = df_val["bias"]
y_test = df_test["bias"]

# 🔹 TF-IDF Vectorization (Reduced `max_features` to avoid memory issues)
vectorizer = TfidfVectorizer(
    ngram_range=(1, 3), stop_words="english",
    max_features=40000, sublinear_tf=True,
    analyzer="word"
)
X_train_final = vectorizer.fit_transform(X_train_text)
X_val_final = vectorizer.transform(X_val_text)
X_test_final = vectorizer.transform(X_test_text)

# 🔹 Apply SMOTETomek for class balancing
class_counts = Counter(y_train)
sampling_strategy = {
    0: int(class_counts[0] * 1.2),  # Oversample class 0 by 20%
    1: int(class_counts[1] * 1.2),  # Oversample class 1 by 20%
    2: int(class_counts[2] * 1.2)   # Oversample class 2 by 20%
}

smote_tomek = SMOTETomek(random_state=42, sampling_strategy=sampling_strategy)
X_train_final, y_train = smote_tomek.fit_resample(X_train_final, y_train)

# 🔹 Train Random Forest Classifier (Further optimized for performance)
model = RandomForestClassifier(n_estimators=300, max_depth=50, random_state=42, class_weight="balanced")
model.fit(X_train_final, y_train)

# Predict and evaluate
y_val_pred = model.predict(X_val_final)
y_test_pred = model.predict(X_test_final)

print("Validation Set Performance:")
print(classification_report(y_val, y_val_pred))

print("Test Set Performance:")
print(classification_report(y_test, y_test_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from scipy.sparse import hstack
from sklearn.metrics import classification_report
from collections import Counter

# Convert lists into DataFrames
df_train = pd.DataFrame(random_train_articles)
df_val = pd.DataFrame(random_val_articles)
df_test = pd.DataFrame(random_test_articles)

# Drop problematic sources
drop_list = {"Guest Writer", "Yellow Scene Magazine", "Aaron Carroll", "Ken Burns",
             "Carol Costello", "The Intelligencer", "Polish Times"}
df_train = df_train.loc[~df_train["source"].isin(drop_list)]
df_val = df_val.loc[~df_val["source"].isin(drop_list)]
df_test = df_test.loc[~df_test["source"].isin(drop_list)]

# Encode 'source' as categorical, handling unseen labels
label_encoder = LabelEncoder()
df_train.loc[:, "source_encoded"] = label_encoder.fit_transform(df_train["source"])
df_val.loc[:, "source_encoded"] = df_val["source"].apply(lambda x: label_encoder.transform([x])[0] if x in label_encoder.classes_ else -1)
df_test.loc[:, "source_encoded"] = df_test["source"].apply(lambda x: label_encoder.transform([x])[0] if x in label_encoder.classes_ else -1)

# Assign bias scores using allsides_media_bias_ratings
df_train.loc[:, "source_bias"] = df_train["source"].map(allsides_media_bias_ratings).fillna(1)
df_val.loc[:, "source_bias"] = df_val["source"].map(allsides_media_bias_ratings).fillna(1)
df_test.loc[:, "source_bias"] = df_test["source"].map(allsides_media_bias_ratings).fillna(1)

# Define features & labels
X_train_text = df_train["content"]
X_val_text = df_val["content"]
X_test_text = df_test["content"]

y_train = df_train["bias"]
y_val = df_val["bias"]
y_test = df_test["bias"]


vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),  # Use only unigrams and bigrams
    stop_words="english",
    max_features=50000,  # Reduce to 50K features
    min_df=10,  # Filter out rare words
    max_df=0.8,  # Exclude overly common words
    sublinear_tf=True
)

X_train_vec = vectorizer.fit_transform(X_train_text)
X_val_vec = vectorizer.transform(X_val_text)
X_test_vec = vectorizer.transform(X_test_text)

# Reduce precision to save RAM
X_train_vec = X_train_vec.astype("float32")
X_val_vec = X_val_vec.astype("float32")
X_test_vec = X_test_vec.astype("float32")

# SMOTE with Lower Sampling
bias_counts = Counter(y_train)
# Limit oversampling to avoid RAM issues
smote = SMOTE(sampling_strategy="auto", random_state=42, k_neighbors=3)
# Apply SMOTE to the combined features to ensure consistency.
X_train_final, y_train = smote.fit_resample(X_train_vec, y_train)

# Train XGBoost for Higher Accuracy
model = SGDClassifier(loss="log_loss", max_iter=1000, alpha=1e-4, penalty="l2", random_state=42)

model.fit(X_train_final, y_train)

# Predictions
# Use X_val_combined and X_test_combined for prediction to maintain consistency.
y_val_pred = model.predict(X_val_vec)
y_test_pred = model.predict(X_test_vec)

# Performance Report
print("Validation Set Performance:")
print(classification_report(y_val, y_val_pred))

print("Test Set Performance:")
print(classification_report(y_test, y_test_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import f_classif, SelectKBest
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
import time

# --- Data Loading ---
df_train = pd.DataFrame(random_train_articles)
df_val = pd.DataFrame(random_val_articles)
df_test = pd.DataFrame(random_test_articles)

# --- Data Cleaning ---
drop_list = {"Guest Writer", "Yellow Scene Magazine", "Aaron Carroll", "Ken Burns",
             "Carol Costello", "The Intelligencer", "Polish Times"}
df_train = df_train.loc[~df_train["source"].isin(drop_list)]
df_val = df_val.loc[~df_val["source"].isin(drop_list)]
df_test = df_test.loc[~df_test["source"].isin(drop_list)]

# --- Features/Labels ---
X_train_text = df_train["content"]
X_val_text = df_val["content"]
X_test_text = df_test["content"]

y_train = df_train["bias"]
y_val = df_val["bias"]
y_test = df_test["bias"]

# Start timer
start_time = time.time()

# 1. Enhanced TF-IDF Vectorization
vectorizer = TfidfVectorizer(
    ngram_range=(1, 4),       # Extended to 4-grams
    max_features=120000,      # Increased feature space
    min_df=5,
    max_df=0.5,               # More aggressive filtering
    sublinear_tf=True,
    stop_words='english',
    analyzer='char_wb'        # Character n-grams for robustness
)

print("Vectorizing text data...")
X_train_vec = vectorizer.fit_transform(X_train_text)
X_val_vec = vectorizer.transform(X_val_text)
X_test_vec = vectorizer.transform(X_test_text)

# 2. Optimized Dimensionality Reduction
print("Reducing dimensions...")
svd = TruncatedSVD(n_components=600, random_state=42)  # Increased components
X_train_svd = svd.fit_transform(X_train_vec)
X_val_svd = svd.transform(X_val_vec)
X_test_svd = svd.transform(X_test_vec)

# 3. Advanced Feature Selection
print("Selecting best features...")
selector = SelectKBest(f_classif, k=50000)  # Keep top 50K features
X_train_selected = selector.fit_transform(X_train_svd, y_train)
X_val_selected = selector.transform(X_val_svd)
X_test_selected = selector.transform(X_test_svd)

# 4. Class Balancing with SMOTE
print("Balancing classes...")
smote = SMOTE(
    sampling_strategy='auto',
    k_neighbors=3,
    random_state=42
)
X_train_final, y_train_final = smote.fit_resample(X_train_selected, y_train)

# 5. Ensemble Model Construction
print("Building ensemble models...")

# Optimized XGBoost
xgb = XGBClassifier(
    n_estimators=1200,
    max_depth=8,
    learning_rate=0.03,
    reg_alpha=8,
    reg_lambda=12,
    gamma=0.5,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method='hist',
    early_stopping_rounds=50,
    eval_metric='merror',
    random_state=42
)

# Enhanced Logistic Regression
logreg = LogisticRegression(
    max_iter=2000,
    class_weight='balanced',
    solver='saga',
    penalty='elasticnet',
    l1_ratio=0.5,
    n_jobs=-1,
    random_state=42
)

# 6. Voting Classifier
ensemble = VotingClassifier(
    estimators=[
        ('xgb', xgb),
        ('lr', logreg)
    ],
    voting='soft',
    n_jobs=-1
)

# 7. Training with Early Stopping
print("Training ensemble model...")

# Create eval set for XGBoost early stopping
eval_set = [(X_val_selected, y_val)]

# Custom wrapper for early stopping
class EarlyStoppingEnsemble:
    def __init__(self, estimators):
        self.estimators = estimators

    def fit(self, X, y):
        for name, est in self.estimators:
            if name == 'xgb':
                est.fit(X, y, eval_set=eval_set, verbose=1)
            else:
                est.fit(X, y)
        return self

    def predict(self, X):
        preds = [est.predict_proba(X) for _, est in self.estimators]
        return np.argmax(np.mean(preds, axis=0), axis=1)

# Train the ensemble
final_model = EarlyStoppingEnsemble([('xgb', xgb), ('lr', logreg)])
final_model.fit(X_train_final, y_train_final)

# 8. Evaluation
print("\n=== Final Model Performance ===")

print("\nValidation Performance:")
y_val_pred = final_model.predict(X_val_selected)
print(classification_report(y_val, y_val_pred, digits=4))

print("\nTest Performance:")
y_test_pred = final_model.predict(X_test_selected)
print(classification_report(y_test, y_test_pred, digits=4))

# Runtime
print(f"\nTotal runtime: {(time.time()-start_time)/60:.1f} minutes")

# 9. Feature Importance Analysis (Optional)
print("\nTop Features:")
feature_names = np.array(vectorizer.get_feature_names_out())[selector.get_support()]
top_features = feature_names[np.argsort(svd.components_[0])[-20:]]
print(top_features)
```
