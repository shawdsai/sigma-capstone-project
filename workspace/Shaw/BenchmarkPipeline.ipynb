```python
import csv
from newspaper import Article
from newspaper import Config
```

```python
def extract_article_info(url):
    config = Config()
    config.browser_user_agent = (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/117.0.0.0 Safari/537.36"
    )
    
    article = Article(url, config=config)
    article.download()
    article.parse()
    
    return article.title, article.text
```

```python
import pandas as pd
import joblib

# === Load saved model and label encoder ===
model = joblib.load("tfidf_xgb_pipeline.joblib")
label_encoder = joblib.load("label_encoder.joblib")

# === Load article data ===
df = pd.read_csv("2024_articles.csv")  # assumes columns: article_id, url, title, content, bias

# === Preprocess text ===
df['title'] = df['title'].fillna('')
df['content'] = df['content'].fillna('')
text_input = df['title'] + " " + df['content']

# === Predict ===
y_pred_encoded = model.predict(text_input)
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# === Extract source from URL ===
df['source'] = df['url'].apply(lambda x: x.split('/')[2].replace("www.", ""))

# === Prepare final output ===
df_output = pd.DataFrame({
    "Source": df['source'],
    "URL": df['url'],
    "Source Leaning": df['bias'],
    "Final Model Label": y_pred  # same as model label here
})

# === Save to CSV ===
df_output.to_csv("2024_articles_predictions.csv", index=False)
print("✅ Predictions saved to 2024_articles_predictions.csv")
```
